```python
import pandas as pd
from nltk import FreqDist
from nltk.tokenize import word_tokenize
```

```python
df =  pd.read_csv('hashtags.csv',  sep=';;', on_bad_lines='skip',lineterminator='\r', engine ='python')
df.head(10)
```

```python
# Tokenize the words
tag_words = df['tag'].apply(word_tokenize)

# Flatten the list of words
tag_words_flat = [word for sublist in tag_words for word in sublist]

# Freq
fdist = FreqDist(tag_words_flat)

# Ordering
most_common_words = fdist.most_common()

for word, count in most_common_words:
    print(f"{word}: {count}")
```
