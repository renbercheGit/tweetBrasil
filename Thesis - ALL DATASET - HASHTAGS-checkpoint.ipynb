Cleaning 1 - Removing rows with specific words (the whole row, because is related to a subject that won't be useful)

```python
# Specify the list of specific words you want to filter out
specific_words = ['trânsito', 'rodovia', 'tráfego', 'ecovias','Anchieta','TIM', 'tim', 'Kaysar' , 'novela', 'Bandeirantes', 'operacaobetalab', 'bbb', 'transito', 'trânsito', 'orcars', 'oscar', 'grammy', 'grammys', 'kaysar', 'timbetalab', ' masterchefbr', 'enem', 'futebol', ' bundesliga', 'corinthians']

# Fill NaN values in the 'text' column with an empty string
df['text'].fillna('', inplace=True)

# Create a boolean mask for rows containing specific words in the 'text' column
mask = df['text'].str.contains('|'.join(specific_words), case=False)

# Apply the mask to filter out rows with specific words
df_filtered = df[~mask]

print("Filtered DataFrame length:", len(df_filtered))
```

```python
# Define a regular expression pattern to match hashtags
hashtag_pattern = r'#\w+'

# Extract hashtags from the 'text' column using regular expressions
hashtags_list = df_filtered['text'].str.findall(hashtag_pattern, flags=re.IGNORECASE)

# Flatten the list of lists
all_hashtags = [tag.lower() for sublist in hashtags_list for tag in sublist]

# Count the occurrences of each hashtag
hashtag_counts = {}
for hashtag in all_hashtags:
    hashtag_counts[hashtag] = hashtag_counts.get(hashtag, 0) + 1

# Organize the hashtag counts in descending order
sorted_hashtag_counts = dict(sorted(hashtag_counts.items(), key=lambda item: item[1], reverse=True))

# in descending order
for hashtag, count in sorted_hashtag_counts.items():
    print(f"{hashtag}: {count}")
```

```python
def parse_date(date_str):
    try:
        return pd.to_datetime(date_str)
    except:
        return pd.NaT  

# Apply the parse_date function to 'created_at' column
df_filtered['created_at'] = df_filtered['created_at'].apply(parse_date)

# Drop rows with missing dates
df_filtered = df_filtered.dropna(subset=['created_at'])

# Extract the 'date' and 'time' components
df_filtered['date'] = df_filtered['created_at'].dt.date
df_filtered['time'] = df_filtered['created_at'].dt.time

# Filter rows for the years 2015 to 2019
dftime_filtered = df_filtered[(df_filtered['created_at'] >= pd.Timestamp('2015-01-01')) & (df_filtered['created_at'] <= pd.Timestamp('2019-12-31'))]

# Group the data by year and count the occurrences
year_frequency = dftime_filtered.groupby(dftime_filtered['created_at'].dt.year)['date'].count()

print(year_frequency)
```

```python
# Drop rows with missing dates
df_filtered = df_filtered.dropna(subset=['created_at'])

# Extract the 'year' and 'month' components from 'created_at'
df_filtered['year'] = df_filtered['created_at'].dt.year
df_filtered['month'] = df_filtered['created_at'].dt.month

# Count the number of hashtags per row
df_filtered['num_hashtags'] = [len(tags) for tags in hashtags_list]

# Filter rows for the years 2015 to 2019
df_filtered_interval = df_filtered[(df_filtered['year'] >= 2015) & (df_filtered['year'] <= 2019)]

# Group the data by year and month, then sum the number of hashtags
monthly_hashtag_counts = df_filtered_interval.groupby(['year', 'month'])['num_hashtags'].sum()

print(monthly_hashtag_counts)
```

```python
from collections import Counter

# Define a regular expression pattern to match hashtags
hashtag_pattern = r'#\w+'

# Extract hashtags from the 'text' column using regular expressions
hashtags_list = df_filtered['text'].str.findall(hashtag_pattern, flags=re.IGNORECASE)

# Apply the parse_date function to 'created_at' column
df_filtered['created_at'] = df_filtered['created_at'].apply(parse_date)

# Drop rows with missing dates
df_filtered = df_filtered.dropna(subset=['created_at'])

# Extract the 'year' and 'month' components from 'created_at'
df_filtered['year'] = df_filtered['created_at'].dt.year
df_filtered['month'] = df_filtered['created_at'].dt.month

# Flatten the list of lists and count hashtag occurrences
all_hashtags = [tag.lower() for sublist in hashtags_list for tag in sublist]
hashtag_counts = Counter(all_hashtags)

# Filter rows for the years 2015 to 2019
df_filtered_interval = df_filtered[(df_filtered['year'] >= 2015) & (df_filtered['year'] <= 2019)]

# Filter rows for the year 2016
df_filtered_2016 = df_filtered_interval[df_filtered_interval['year'] == 2016]

# Group the data by year, month, and hashtag
monthly_hashtags = df_filtered_2016.groupby(['year', 'month'])['text'].apply(lambda x: [hashtag.lower() for sublist in x.str.findall(hashtag_pattern, flags=re.IGNORECASE) for hashtag in sublist])

# Count the occurrences of each hashtag per month
monthly_hashtag_counts = monthly_hashtags.apply(Counter)

# Get the top 30 hashtags by month
top_30_hashtags = monthly_hashtag_counts.apply(lambda x: x.most_common(30))

# Print the top 30 hashtags for each month
for year_month, hashtags in top_30_hashtags.items():
    print(f"Top 30 hashtags for {year_month}:")
    for hashtag, count in hashtags:
        print(f"{hashtag}: {count}")
    print()
```
